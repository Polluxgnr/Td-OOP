##OperatorOverloading


###Problem1

```python
class Complex:
    def __init__(self, real, imaginary):
        self.real=real
        self.imaginary=imaginary

    def __add__(self, other):
        return Complex(self.real+other.real, self.imaginary+other.imaginary)

    def __sub__(self, other):
        return Complex(self.real-other.real, self.imaginary-other.imaginary)

    def __repr__(self):
        return f"{self.real}+{self.imaginary}i"
```

#### test

```python
num1=Complex(4, 2)
num2=Complex(6, 3)
print(f"Addition: {num1}+{num2}={num1+num2}")
print(f"Subtraction: {num1}-{num2}={num1-num2}")
```

```python
class Vector:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def __mul__(self, other):
        return (self.x * other.x) + (self.y * other.y) + (self.z * other.z)

    def __repr__(self):
        return f"Vector({self.x},{self.y},{self.z})"
```

###Problem5

```python
class Circle:
    def __init__(self, radius):
        self.radius=radius

    def __lt__(self, other):
        return self.radius<other.radius

    def __gt__(self, other):
        return self.radius>other.radius

    def __le__(self, other):
        return self.radius<=other.radius

    def __ge__(self, other):
        return self.radius>=other.radius

    def __eq__(self, other):
        return self.radius==other.radius

    def __ne__(self, other):
        return self.radius!=other.radius
```

```python
#Testing all circle boundary conditions
c1=Circle(10)
c2=Circle(5)
print(f"c1 > c2: {c1>c2}")
print(f"c1 == c2: {c1==c2}")
print(f"c1 <= c2: {c1<=c2}")
```

###Problem6

```python
import math

class Euclidean:
    def __init__(self, x, y):
        self.x=x
        self.y=y

    def distance(self):
        return math.sqrt(self.x**2+self.y**2)

    def __lt__(self, other):
        return self.distance()<other.distance()

    def __gt__(self, other):
        return self.distance()>other.distance()

    def __le__(self, other):
        return self.distance()<=other.distance()

    def __ge__(self, other):
        return self.distance()>=other.distance()

    def __eq__(self, other):
        return self.distance()==other.distance()

    def __ne__(self, other):
        return self.distance()!=other.distance()
```

```python

p1=Euclidean(3, 4) #Dist = 5.0
p2=Euclidean(1, 1) #Dist ~ 1.41
print(f"Point 1 Distance: {p1.distance():.2f}")
print(f"Is p1 further than p2? {p1 > p2}")
```

###Problem7

```python
class Matrix:
    def __init__(self, data):
        self.data=data
        self.rows=len(data)
        self.cols=len(data[0])

    def __mul__(self, other):
        result=[[0 for _ in range(other.cols)] for _ in range(self.rows)]

        for i in range(self.rows):
            for j in range(other.cols):
                for k in range(self.cols):
                    result[i][j]+=self.data[i][k]*other.data[k][j]

        return Matrix(result)

    def __repr__(self):
        return "\n".join([str(row) for row in self.data])
```

```python
#2x2 Matrix multiplication test
m1=Matrix([[1, 2], [3, 4]])
m2=Matrix([[2, 0], [1, 2]])
print("M1*M2:")
print(m1*m2)
```

```python
class Vehicle:
    def drive(self):
        return "Driving a vehicle"


class Car(Vehicle):
    def drive(self):
        return "Driving a car"


class Bicycle(Vehicle):
    def drive(self):
        return "Riding a bicycle"
```

###Problem2

```python
class Person:
    def __init__(self, name, age, address):
        self.name=name
        self.age=age
        self.address=address

    def introduce(self):
        return f"Hi, I'm {self.name}, {self.age} years old, living at {self.address}."


class Student(Person):
    def __init__(self, name, age, address, field_of_study):
        super().__init__(name, age, address)
        self.field_of_study=field_of_study

    def introduce(self):
        return f"{super().introduce()} I study {self.field_of_study}."


class Employee(Person):
    def __init__(self, name, age, address, company):
        super().__init__(name, age, address)
        self.company=company

    def introduce(self):
        return f"{super().introduce()} I work at {self.company}."
```

```python
#Testing personalized introductions
student=Student("Alice", 21, "123 University Ave", "Computer Science")
employee=Employee("Bob", 35, "456 Corporate Blvd", "Google")

print(student.introduce())
print(employee.introduce())
```

###Problem3

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number=account_number
        self.balance=balance

    def deposit(self, amount):
        self.balance+=amount

    def withdraw(self, amount):
        if amount<=self.balance:
            self.balance-=amount
            return True
        return False

    def transfer(self, destination_account, amount):
        if self.withdraw(amount):
            destination_account.deposit(amount)
            return True
        return False


class SavingsAccount(BankAccount):
    def __init__(self, account_number, balance, interest_rate):
        super().__init__(account_number, balance)
        self.interest_rate=interest_rate

    def add_interest(self):
        interest=self.balance*self.interest_rate
        self.deposit(interest)

    def check_balance(self):
        return f"Account {self.account_number} Balance: {self.balance}"
```

Test

```python
#Setup accounts
personal_acc = BankAccount("MAIN-404", 1000)
savings_acc = SavingsAccount("SAVE-999", 500, 0.05) # 5% interest

print(f"Initial Main Balance: {personal_acc.balance}€")

#Attempting a transfer
if personal_acc.transfer(savings_acc, 200):
    print("Transfer successful: 200€ moved to Savings.")
else:
    print("Transfer failed: Insufficient funds.")

#Applying annual interest
savings_acc.add_interest()

print(f"Main Account remaining: {personal_acc.balance}€")
print(savings_acc.check_balance())
```
